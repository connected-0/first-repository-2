```python

from random import randrange
import random

class FunnyDice:
    def __init__(self, n=6): 
        
        self.n = int(n)
        self.numbers = [0] * n
        
    def throw(self):
        
        for i in range(6):
            self.index = randrange(0,self.n)
            self.numbers[self.index] += 1   
        

    def getval(self):
        return self.numbers 


def get_inputs():
    n = int(input("주사위 면의 개수를 입력하세요: "))
    return n 

def main(): 
    n = get_inputs() 
    mydice = FunnyDice(n)
    mydice.throw()
    
    diceNum = mydice.getval()
    maxNum = max(diceNum)
    answer = [index + 1 for index, val in enumerate(diceNum) if val == maxNum]
    for i in range(n):
        print("{}은 {}번 나왔습니다.".format(i + 1, diceNum[i]))
    print("당신의 행운의 숫자는 {}".format(answer))

if __name__ == '__main__':
    main()
```

```python
from random import randrange
import random

class FunnyDice:
    def __init__(self, n=6): # 매 번 바뀌는 주소값을 바인딩하기 위해서 self를 사용한다

        self.n = int(n)
        self.numbers = list(range(1, n+1)) 
        self.index = randrange(0, self.n) 
        self.val = self.numbers[self.index] 

    def throw(self):

        self.index = randrange(0,self.n)
        self.val = self.numbers[self.index] # 하나 뽑는 값 지정

    def getval(self):
        return self.val # 뽑은 값 출력

    def dice_tot(self,a=6):
        result = [0] * self.n
        for i in range(a):
            dice=random.randint(1, self.n)
            result[dice-1] += 1
        return result    

        
def get_inputs():
    n = int(input("주사위 면의 개수를 입력하세요: "))
    a = int(input("주사위 모드를 선택하시오\n1.일반주사위 던지기 \n2.주사위의 통계\n"))
    if a == 2 :
        b = int(input("주사위를 던질 수를 정하시오."))
        return n, a, b    
    else :
        return n, a 
    

def main(): 
    n = get_inputs() 
    if n[1]==1:
        mydice = FunnyDice(n[0])
        mydice.throw()
        print("행운의 숫자는? {0}".format(mydice.getval()))
    else :
        mydice = FunnyDice(n[0])
        list_tot= mydice.dice_tot(n[2])
        max_num = max(list_tot)
        answer = [num + 1 for num, value in enumerate(list_tot) if value == max_num]
        for i in range(n[0]) :
            print(i+1 ,"은" ,list_tot[i],"번 나왓습니다.")
        print("행운의 숫자는",answer )


if __name__ == '__main__':
    main()
```
